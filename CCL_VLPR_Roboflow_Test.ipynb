```python
import os
import json
from PIL import Image
from skimage import io

from roboflow import Roboflow
rf = Roboflow(api_key="XEsK3U8yRnvoLkCHX9qE")
project = rf.workspace().project("ocr-vehicle-licence-plate.")
model = project.version(3).model



folder_path = 'Images'

# loop through all the files in the folder
for file_name in os.listdir(folder_path):
    # check if the file has the .jpg extension
    if file_name.endswith('.jpg'):
        # print(file_name)
        lp_image = 'Images/' + file_name
        img = io.imread(lp_image)

        # infer on a local image
        json_string = model.predict(lp_image, confidence=10, overlap=30).json()
        # print(json_string)
        # json_string

        # json_data = json.loads(json_string)
        sorted_predictions = sorted(json_string['predictions'], key=lambda x: x['x'])
        recognized = []

        for platenum in sorted_predictions:
          recognized.append(platenum['class'])

        print(file_name)
        print(recognized)
        print()

        io.imshow(img)






# visualize your prediction
# model.predict("your_image.jpg", confidence=40, overlap=30).save("prediction.jpg")

# infer on an image hosted elsewhere
# print(model.predict("URL_OF_YOUR_IMAGE", hosted=True, confidence=40, overlap=30).json())
```

```python
import matplotlib.pyplot as plt

folder_path = 'Images'

# get the number of jpg files in the folder
num_images = len([file_name for file_name in os.listdir(folder_path) if file_name.endswith('.jpg')])

# create a figure with subplots
fig, axs = plt.subplots(1, num_images)

# loop through all the files in the folder
for idx, file_name in enumerate(os.listdir(folder_path)):
    # check if the file has the .jpg extension
    if file_name.endswith('.jpg'):
        # read the image
        img = plt.imread(os.path.join(folder_path, file_name))
        # display the image in a subplot
        axs[idx].imshow(img)

plt.show()


```
